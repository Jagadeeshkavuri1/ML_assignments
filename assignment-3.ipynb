#1) Basic Line Plot

##(i)Plot a simple line graph of ùë¶ = ùë•^2for ùë• values ranging from -10 to 10.



```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10)
y = x**2
plt.plot(x,y)
plt.show()
```

##(ii)Label the x-axis as "X Values" and the y-axis as "Squared Values."


```python
plt.xlabel("X Values")
plt.ylabel("Squared Values.")
plt.plot(x,y)
plt.show()
```

##(iii)Add a title to the graph.

```python
plt.title("squared graph")
plt.plot(x,y)
plt.show()
```

#(2). Bar Plot

##Create a bar chart showing the marks scored by 5 students: [Sai, 'Vamsi','Divij', 'Sritan', 'Sita'] with marks [85, 90, 78, 92, 88].


```python
students = ['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks = [85, 90, 78, 92, 88]
print("student names",students)
print("marks of the tsudent",marks)
plt.bar(students,marks)
plt.show()
```

##(ii) Add labels to the x-axis and y-axis and give the chart a title.

```python
plt.xlabel('Students')
plt.ylabel('Marks')
plt.bar(students,marks)
plt.show()
```
