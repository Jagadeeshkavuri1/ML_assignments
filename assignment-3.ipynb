#1) Basic Line Plot

##(i)Plot a simple line graph of ùë¶ = ùë•^2for ùë• values ranging from -10 to 10.



```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10)
y = x**2
plt.plot(x,y)
plt.show()
```

##(ii)Label the x-axis as "X Values" and the y-axis as "Squared Values."


```python
plt.xlabel("X Values")
plt.ylabel("Squared Values.")
plt.plot(x,y)
plt.show()
```

##(iii)Add a title to the graph.

```python
plt.title("squared graph")
plt.plot(x,y)
plt.show()
```

#(3)Scatter Plot

##(i)Generate two lists of 20 random integers each (for x and y coordinates).

```python
import random
x_coordinates = [random.randint(0, 100) for _ in range(20)]
y_coordinates = [random.randint(0, 100) for _ in range(20)]
print("the two list are  ")
print("X Coordinates:", x_coordinates)
print("Y Coordinates:", y_coordinates)
```

##(ii) o Create a scatter plot with these values, and customize it by adding color,size, and labels to the points.


```python
plt.scatter(x_coordinates,y_coordinates,c='red',s=35,alpha=0.5)
labels = ['A', 'B', 'C', 'D', 'E','f','g','h','i','j','k','l','m','n','o','p','q','r','y','w']
for i in range(len(x_coordinates)):
    plt.annotate(labels[i], (x_coordinates[i], y_coordinates[i]),textcoords="offset points", xytext=(5, 5))
plt.title("scatter graph")
plt.xlabel("x-aixs")
plt.ylabel("y-axis")
plt.show()
```

#(5). Pie Chart

##(i)o Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E']with values [20, 30, 25, 15, 10].

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
plt.pie(values, labels=categories)
plt.show()
```

##(ii)o Display the percentage values on the chart and use a color palette fordifferentiation.

```python
col = plt.cm.Paired(np.linspace(0, 1, len(categories)))
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=col)
plt.show()
```

#(6). Box Plot (Seaborn)

##(i) Create a box plot to visualize the distribution of the marks scored bystudents in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73].

```python
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]
plt.boxplot([78, 85, 92, 88, 76, 95, 89, 84, 91, 73])
plt.ylabel('Marks')
plt.show()
```

##(ii)Highlight the median and any outliers.

```python
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]
plt.boxplot(marks,
            flierprops=dict(marker='o', color='yellow', markersize=25),
            medianprops=dict(color='red', linewidth=2))
plt.ylabel('Marks')
plt.show()
```

##(i)Generate a pair of plots (bar charts or line plots) using the same data inboth Matplotlib and Seaborn.

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]

plt.subplot(1, 2, 1)
plt.plot(categories, values, marker='o', color='green')
plt.title('Line Plot (Matplotlib)')

plt.subplot(1, 2, 2)
sns.lineplot(x=categories, y=values, marker='o', color='red')
plt.title('Line Plot (Seaborn)')

plt.tight_layout()
plt.show()
```
