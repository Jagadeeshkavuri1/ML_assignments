**this is the histogram dta visualization method**

```python
sns.set(style="whitegrid")
plt.figure(figsize =(10,10))
sns.histplot(data['texture_worst'],bins=11,color='red',edgecolor='black')
plt.title('histogram of teture worst')
plt.xlabel('texture worst ')
plt.ylabel('count')
plt.show()
```

**this is the boxplot data visualization method**

```python
plt.figure(figsize=(10,10))
sns.boxplot(data['texture_worst'])
plt.title('boxplot of texture worst')
plt.xlabel('texture worst')
plt.show()
```

**this is the barchart data visualization method**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data2=data[['diagnosis', 'texture_mean', 'texture_worst']].melt(id_vars='diagnosis'),
            x='variable', y='value', hue='diagnosis')
plt.title('Boxplot of Texture Mean and Texture Worst by Diagnosis')
plt.xlabel('Texture Features')
plt.ylabel('Value')
plt.legend(title='Diagnosis')
plt.grid(True)
plt.show()
```
